# Introduction :

<h2> 1) Meaning of open cluster </h2>
<p> The meaning of 'open cluster' in layman terms is : A loose grouping of stars. </p>
<h2> 2) Spectral classification of stars </h2>
<img src = "http://planetfacts.org/wp-content/uploads/2011/04/Spectral-class-stars.jpg">
<p style="text-align:center; font-size:10px">Image Source: http://planetfacts.org/wp-content/uploads/2011/04/Spectral-class-stars.jpg</p>
<p>Stars are classified by their spectra (the elements that they absorb) and their temperature. There are seven main types of stars. In order of decreasing temperature: O, B, A, F, G, K, and M.

O and B stars are uncommon but very bright; M stars are common but dim.
In the present classification system, the letter is followed by a number from 0 to 9. The bigger is the number, lower is the temperature of the star. Ex: An A0 type star is hotter than an A3 type star. 
    
Further, the colour of a star is related to its temperature. From O type to M type stars, the colours change from blue to red as the temperatures of the stars reduce.  
  
</p>
<h2> 3) Meaning of Colour index </h2>
<p> Colour index is a simple numerical expression that determines the colour of an object, which in the case of a star gives its temperature.
B-V, U-B, V-R, etc. are some examples of colour indexes. Lower the value, hotter is the star (bluer is colour).
Here B, V, U represent the magnitudes of the star as seen in Blue, Visible and ultraviolet filters respectively.
</p>
<h2> 4) Magnitudes </h2>
<p> Apparent magnitudes measure how bright or faint something appears from our perspective on the Earth. The magnitude scale is a logarithmic scale which actually measures the radiative flux of the star's light that arrives on the Earth. 
</p>   
<b> m = m(reference) - 2.5*log(f) </b>
<p> Here m, f are apparent magnitude and flux respectively. </p>


#  Description Of HR Diagram :

<p>
    The Hertzsprung–Russell diagram, abbreviated as H–R diagram, HR diagram or HRD, is a scatter plot of stars showing the relationship between the stars' absolute magnitudes or luminosities versus their stellar classifications or effective temperatures.
</p>
<p>
    Thus, the horizontal axis in a HR diagram can be effective temperature, colour indices or
spectral class, while the vertical axis can be luminosity with respect to that of the Sun or
the absolute magnitude MV.
</p>
<img src = "https://astronomy.swin.edu.au/cms/cpg15x/albums/userpics/hrdiagram1.jpg">
<p style="text-align:center; font-size:10px">Image Source: https://astronomy.swin.edu.au/cms/cpg15x/albums/userpics/hrdiagram1.jpg</p>
<p>
    When luminosity is plotted as a function of the temperature for a large number of stars,
stars do not fall randomly on the graph; rather they are confined to specific regions. This
tells you that there is some physical relationship between the luminosity and temperature
of a star. From the figure, one sees that most stars fall along a diagonal strip from high
temperature, high luminosity stars to low temperature, low luminosity stars. These are
    the <b>main sequence stars</b>. Our Sun is one of them. There are a few stars that are not in this
diagonal strip. There are some low temperature, high luminosity stars ‐ these are called
    <b>giants and supergiants</b>. The reason they are so luminous while being relatively cool is
because they are so big (50 times more massive than our Sun). Another group of stars are
in the high temperature, low luminosity corner of the diagram. Since these stars are hot,
    but not very luminous, they must be very small, so they're called <b>white dwarfs</b>.
</p>

# Brief description of different types of stars:

<h3>The Luminosity – Radius – Temperature relation</h3>
<p>
    The Luminosity of a star is proportional to its Effective Temperature to the 4 power and
its Radius squared.
</p>
<p>
    L = 4πR<sup>2</sup>σT<sup>4</sup>

Here, σ is the Boltzmann's constant.
</p>
<h2> 1) Main sequence stars </h2>
<p>Most nearby stars (85%), including the Sun, lie along a diagonal band in the H‐R diagram
called the Main Sequence.
</p>
<p>
Ranges of stellar properties:
<p>
    • L=10<sup>‐2</sup> to 10<sup>6</sup> Lsun
</p>
<p>   
• T=3000 to >50,000 K
</p>
<p>
• R=0.1 to 10 Rsun
</p>
<h4> The mass - luminosity relation : </h4>
<p>
    The brightness of
Main‐Sequence stars varies proportionally to
some power of their masses. For most of
the range of stellar masses, the
proportionality is as the 3.5 power of the
mass.
</p>
<h4> Mass - Lifetime Relation of Main Sequence Stars </h4>
<p>A star's lifetime on the main sequence is how long it takes to use up the hydrogen in its
core. The luminosity (L) of a star is a measure of how rapidly it is using up its hydrogen.
The mass (M) of a star is a measure of how much fuel it has. The time it takes to use up
the fuel is proportional to its amount of fuel (M) divided by the rate of fuel consumption
(L)
</p>
<b>Since L ~ M<sup>3.5</sup> (mass - luminosity relation)

Hence, t ~ M/L ~ M/(M<sup>3.5</sup>) ~ 1/M<sup>2.5</sup></b>
<h2> 2) Giants & Supergiants </h2>
<p>
    There are also two bands of stars in the H‐R diagram that are brighter than Main
Sequence stars with the same effective temperatures.
</p>
<p>
    <b>Giants</b>: Large but cool stars with a wide range of luminosities:
</p>
<p>
• R = 10 to 100 Rsun
</p>
<p>
• L = 10<sup>3</sup> to 10<sup>5</sup> Lsun
</p>
<p>
    <b>Supergiants</b>: The very largest stars, arranged along the top of the H‐R diagram with
a wide range of effective temperatures but relatively narrow range of
luminosities:
</p>
<p>
• R > 10<sup>3</sup> Rsun
</p>
<p>
• L = 10<sup>5</sup> to 10<sup>6</sup> Lsun
</p>
<h2> 3) White Dwarfs </h2>
<p>There are also a few very hot but also very faint stars that occupy the lower left‐hand
corner of the H‐R Diagram. These are stars that are much fainter than Main Sequence
stars of the same temperature.
</p>
<p>
    Using the Luminosity‐Radius‐Temperature relation, we can make a prediction:
</p>
<p>
• R ~ 0.01 Rsun
</p>
<p>
This is about the size of the Earth!
</p>

# Creating a colour‐magnitude diagram of NGC - 6913.

```python
# Importing libraries.
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
# Importing all the given files.
MB = fits.open('MasterBias.fit')
MB_data = MB[0].data
NB = fits.open('NormFlatB.fit ')
NB_data = NB[0].data
NV = fits.open('NormFlatV.fit')
NV_data = NV[0].data
B10 = fits.open('M29_B_ExpTime10s.fit')
B10_data = B10[0].data
B30 = fits.open('M29_B_ExpTime30s.fit')
B30_data = B30[0].data
B60 = fits.open('M29_B_ExpTime60s.fit')
B60_data = B60[0].data
V10 = fits.open('M29_V_ExpTime10s.fit')
V10_data = V10[0].data
V30 = fits.open('M29_V_ExpTime30s.fit')
V30_data = V30[0].data
V60 = fits.open('M29_V_ExpTime60s.fit')
V60_data = V60[0].data
```

```python
# Subtracting bias from all images by mean bias frame.
B10_data = B10_data - MB_data
B30_data = B30_data - MB_data
B60_data = B60_data - MB_data
V10_data = V10_data - MB_data
V30_data = V30_data - MB_data
V60_data = V60_data - MB_data
```

```python
# Dividing the resulting images by the scaled flat-field image.
# For B
B10_data = B10_data / NB_data
B30_data = B30_data / NB_data
B60_data = B60_data / NB_data
# For V
V10_data = V10_data / NV_data
V30_data = V30_data / NV_data
V60_data = V60_data / NV_data
```

```python
# Averaging the resulting images of 10s, 30s and 60s of exposure time.
B_data = (B10_data + B30_data + B60_data) / 100
V_data = (V10_data + V30_data + V60_data) / 100
```

```python
# Viewing the image in B filter.
# Having a dark background is advisable.
plt.style.use('dark_background')
plt.imshow(B_data, cmap = 'gray')
```

```python
# Viewing the image in V filter.
plt.imshow(V_data, cmap = 'gray')
```

```python
# Logarithmic normalization is used to enhance dark areas by expanding their range, while not clipping bright areas.
# It is done by taking log(image +1).
# Image in V filter.
plt.imshow(np.log(V_data + 1), cmap = 'gray')
```

```python
# Image in B filter.
plt.imshow(np.log(B_data + 1), cmap = 'gray')
```

```python
# A function is to be defined such that the locations of n (user input) most bright pixels (stars) are stored in
# an array and returned. The inputs of the function are a numpy image array, n (number of bright stars to be found) and
# spread (number of pixels) which ensures that the same number of pixels neighbouring a bright pixel are set to 0 
# so that the pixels corresponding to the same star aren't considered to be different stars.
def n_brightest_stars(array, n, spread):
    arr = array.copy()
    location = np.zeros((n,2))
    location = location.astype('int64')
    for i in range(0, n):
        location[i] = np.argwhere(arr == arr.max())
        arr[location[i,0] - spread : location[i,0] + spread, location[i,1] - spread : location[i,1] + spread] = 0
    return location
```

```python
# Messier 29 has a total of 50 stars.
# Let us take 30 stars and let spread be 50 pixels.
location = n_brightest_stars(V_data, 30, 50)
location
```

```python
# Encircling 30 brightest stars in the image.
fig, ax = plt.subplots() 
plt.imshow(np.log(V_data + 1), cmap = 'gray' )
for loc in location:
    x2,x1 = loc
    marker = plt.Circle((x1, x2), radius = 30, color='cyan',linewidth = 0.5 ,fill = False)
    ax.add_patch(marker)
```

```python
# Verifying with image of B filter.
fig, ax = plt.subplots() 
plt.imshow(np.log(B_data + 1), cmap = 'gray' )
for loc in location:
    x2,x1 = loc
    marker = plt.Circle((x1, x2), radius = 30, color='cyan',linewidth = 0.5 ,fill = False)
    ax.add_patch(marker)
```

```python
# A function to find out the B-V, V magnitudes based on the magnitude transformation equation given in the introduction 
# which relates apparent magnitude and gray scale intensity of a pixel assuming a reference value for the magnitude.
# The function takes the reference magnitudes of both B and V as input.
def convert_to_magnitude(ref_B, ref_V):
    mag_B = []
    mag_V = []
    for i in range(0, location.shape[0]):
        mag_B.append(ref_B - 2.5*np.log10(B_data[location[i,0], location[i,1]]))
        mag_V.append(ref_V - 2.5*np.log10(V_data[location[i,0], location[i,1]]))
    B_V = np.array(mag_B) - np.array(mag_V)
    return B_V, mag_B, mag_V
```

```python
# The apparent visual magnitude of Messier 29 is 7.1. Hence, the reference magnitudes are adjusted for most stars to 
# have their visual magnitudes near 7.
B_V, mag_B, mag_V = convert_to_magnitude(11, 12.5)
plt.scatter(B_V, mag_V)
plt.ylim(np.array(mag_V).max() + 1, np.array(mag_V).min() - 1)
plt.xlim(np.array(B_V).min() - 3, np.array(B_V).max() + 1)
plt.xlabel("B-V magnitude")
plt.ylabel("V Magnitude")
plt.title("Colour Magnitude Diagram")
```
